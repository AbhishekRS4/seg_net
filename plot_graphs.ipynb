```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
metrics_nn = np.load('model_nn_100/metrics.npy').item()
metrics_bi = np.load('model_bilinear_100/metrics.npy').item()
metrics_tr = np.load('model_tr_conv_100/metrics.npy').item()
metrics_ba = np.load('model_bayesian_100/metrics.npy').item()
```

```python
x_values = np.arange(1, len(metrics_nn['train_loss']) + 1)

plt.figure(figsize = (12, 8))
plt.title('loss vs epoch')
plt.xlabel('epoch')
plt.ylabel('loss')

plt.plot(x_values, metrics_nn['train_loss'], 'b*-', label = 'nearest neighbor training loss')
plt.plot(x_values, metrics_nn['valid_loss'], 'b--', label = 'nearest neighbor validation loss')

plt.plot(x_values, metrics_bi['train_loss'], 'g*-', label = 'bilinear training loss')
plt.plot(x_values, metrics_bi['valid_loss'], 'g--', label = 'bilinear validation loss')

plt.plot(x_values, metrics_tr['train_loss'], 'm*-', label = 'transposed conv training loss')
plt.plot(x_values, metrics_tr['valid_loss'], 'm--', label = 'transposed conv validation loss')

plt.plot(x_values, metrics_ba['train_loss'], 'y*-', label = 'bayesian training loss')
plt.plot(x_values, metrics_ba['valid_loss'], 'y--', label = 'bayesian validation loss')


plt.legend()
plt.show()
```

```python
plt.figure(figsize = (12, 8))
plt.title('accuracy vs epoch')
plt.xlabel('epoch')
plt.ylabel('accuracy')

plt.plot(x_values, metrics_nn['valid_acc'], 'b--', label = 'nearest neighbor validation accuracy')
plt.plot(x_values, metrics_bi['valid_acc'], 'g--', label = 'bilinear validation accuracy')
plt.plot(x_values, metrics_tr['valid_acc'], 'm--', label = 'transposed conv validation accuracy')
plt.plot(x_values, metrics_ba['valid_acc'], 'y--', label = 'bayesian validation accuracy')

plt.legend()
plt.show()
```
